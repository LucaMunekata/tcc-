```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import folium

ciclovias = gpd.read_file('Estrutura ciclovi√°ria SP/Ciclovias.shp')
ciclovias.head()
```

```python
ciclovias.plot()
plt.show()
```

```python
print(ciclovias.crs)
```

```python
ciclovias.crs = 'epsg:31983'
ciclovias = ciclovias.to_crs(epsg = 4326)

print(ciclovias.crs)
```

```python
map_sp = folium.Map(location = [-23.5789, -46.6388], tiles = "CartoDB Positron", zoom_start = 10)
folium.GeoJson(ciclovias,
        style_function = lambda feature: {
        'weight': 3,
        'color': 'red'
        },
        name = 'ciclovias'
    ).add_to(map_sp)
folium.LayerControl().add_to(map_sp)
map_sp
```
